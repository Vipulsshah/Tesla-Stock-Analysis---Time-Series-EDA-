```python
import pandas as pd
import pandas_datareader as pdr
import datetime as datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_tesla = pdr.get_data_yahoo('TSLA')
```

```python
df_tesla.head()
```

```python
df_tesla['Open'].plot(figsize=(15,6))
```

### 1) Simple Moving Average:

```python
df_tesla['High'].rolling(5).mean().plot()
```

```python
df_tesla['21 day High SMA'] = df_tesla['High'].rolling(window=21,min_periods=1).mean()
```

```python
df_tesla['50 day High SMA'] = df_tesla['High'].rolling(window=50,min_periods=1).mean()
df_tesla['100 day High SMA'] = df_tesla['High'].rolling(window=100,min_periods=1).mean()
```

```python
df_tesla[['High','21 day High SMA']].plot(xlim=['2019-01-01','2022-03-30'],figsize=(15,5))
```

```python
df_tesla[['High','21 day High SMA','50 day High SMA','100 day High SMA']].plot(xlim=['2020-01-01','2022-03-30'],figsize=(15,5))
```

### 2) Cumilative Moving Avg:

```python
df_tesla['High'].expanding().mean().plot(figsize=(15,5))
```

### 3) Exponential weighted moving avg:

```python
df_tesla['EMA 21 days'] = df_tesla['Open'].ewm(span=21,adjust=False).mean()
df_tesla['EMA 50 days'] = df_tesla['Open'].ewm(span=50,adjust=False).mean()
df_tesla['EMA 100 days'] = df_tesla['Open'].ewm(span=100,adjust=False).mean()
```

```python
df_tesla[['Open','EMA 21 days']].plot(xlim=['2020-01-01','2022-3-30'],figsize=(15,5))
```

```python
df_tesla[['Open','EMA 21 days','EMA 50 days','EMA 100 days']].plot(xlim=['2021-01-01','2022-3-30'],figsize=(15,5))
```
